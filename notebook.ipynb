```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import imageio
```

```python
carpark_framethreshold_path = "assets/carpark_framethreshold.png"
carpark_framedilate_path = "assets/carpark_framedilate.png"
base_image_path = (
    "assets/image.png"  # This will be displayed with the title "Base Image"
)
slot_count_path = "assets/slot_count.png"

carpark_framethreshold = imageio.imread(carpark_framethreshold_path)
carpark_framedilate = imageio.imread(carpark_framedilate_path)
base_image = imageio.imread(base_image_path)
slot_count = imageio.imread(slot_count_path)
```

```python
fig, axs = plt.subplots(2, 2, figsize=(15, 10))

axs[0, 0].imshow(carpark_framethreshold)
axs[0, 0].set_title("Carpark Frame Threshold")
axs[0, 0].axis("off")

axs[0, 1].imshow(carpark_framedilate)
axs[0, 1].set_title("Carpark Frame Dilate")
axs[0, 1].axis("off")

axs[1, 0].imshow(base_image)
axs[1, 0].set_title("Base Image")
axs[1, 0].axis("off")

axs[1, 1].imshow(slot_count)
axs[1, 1].set_title("Slot Count")
axs[1, 1].axis("off")

plt.tight_layout()
plt.show()
```
